```python
import json

# 파일 경로 설정
file_path = 'NLP_HyperClovax_for_training_on_colab.ipynb' # 오류가 난 파일명으로 변경

# 파일 읽기
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 문제가 되는 'widgets' 키가 metadata 안에 있다면 삭제
if 'widgets' in data.get('metadata', {}):
    del data['metadata']['widgets']
    print("metadata.widgets has been removed.")
else:
    print("No widgets metadata found.")

# 파일 덮어쓰기 (저장)
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=1) # indent=1은 파일 용량을 줄이기 위함이나 가독성을 위해 2~4로 해도 무방

print("Done. You can now upload this file to GitHub.")
```
