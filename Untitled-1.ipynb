```python
import json
import time
import pandas as pd
import google.generativeai as genai
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# 1. API ì„¤ì • (ë³¸ì¸ì˜ API KEYë¥¼ ë„£ì–´ì£¼ì„¸ìš”)
# ==============================================================================
API_KEY = "AIzaSyD9WF_Z8bkAMP5zc14JbTPkIDOlKVVmR3Y"  # <-- ì—¬ê¸°ì— í‚¤ ì…ë ¥
genai.configure(api_key=API_KEY)

# í‰ê°€ì—ëŠ” ë…¼ë¦¬ì  ì¶”ë¡ ì´ ê°•í•œ Pro ëª¨ë¸ ì‚¬ìš© ê¶Œì¥
model = genai.GenerativeModel('gemini-2.0-flash')

# ==============================================================================
# 2. í‰ê°€ í”„ë¡¬í”„íŠ¸ ì„¤ê³„ (System Prompt)
# ==============================================================================
# ë£¨ë¸Œë¦­ì„ ëª…í™•í•˜ê²Œ ì •ì˜í•˜ì—¬ ì ìˆ˜ì˜ ì¼ê´€ì„±ì„ í™•ë³´í•©ë‹ˆë‹¤.
JUDGE_PROMPT = """
You are an expert linguist acting as a judge for a Korean-English translation contest.
Your task is to evaluate the 'Candidate Translation' based on the 'Source Text' and the 'Reference Translation'.

### Evaluation Rubric (0-5 Scale)

1. **Semantic Accuracy (ì˜ë¯¸ì  ì •í™•ì„±)**
   - 0: Completely wrong meaning or hallucination.
   - 5: Perfect meaning preservation without any distortion.

2. **Syntactic Fluency (í†µì‚¬ì  ìœ ì°½ì„±)**
   - 0: Broken grammar, unintelligible (Word salad).
   - 5: Flawless, native-level natural Korean phrasing.

3. **Cultural Resonance (ë¬¸í™”ì  ê³µëª…)**
   - 0: Literal translation that sounds awkward or foreign.
   - 5: Perfectly localized using appropriate slang/memes/tone (e.g., using "ì‹œë§ˆì´", "ë¬´í•œ ìŠ¤í¬ë¡¤ë§").

4. **Style & Formality (ë¬¸ì²´ ë° ê²©ì‹ ì¶©ì‹¤ë„)**
   - 0: Tone mismatch (e.g., using polite speech when casual is needed).
   - 5: Tone perfectly matches the source's intent (e.g., sarcastic, casual, rude).

### Input Data
- **Source:** {source}
- **Reference:** {reference}
- **Candidate:** {candidate}

### Output Format (Strict JSON)
Provide the output in the following JSON structure ONLY:
{{
  "semantic_score": <int 0-5>,
  "fluency_score": <int 0-5>,
  "cultural_score": <int 0-5>,
  "style_score": <int 0-5>,
  "reasoning": "<Short explanation in Korean describing why these scores were given>"
}}
"""

# ==============================================================================
# 3. í‰ê°€ í•¨ìˆ˜
# ==============================================================================
def evaluate_translation(source, reference, candidate):
    try:
        # í”„ë¡¬í”„íŠ¸ í¬ë§·íŒ…
        prompt = JUDGE_PROMPT.format(
            source=source,
            reference=reference,
            candidate=candidate
        )
        
        # API í˜¸ì¶œ (temperature=0.0ìœ¼ë¡œ ì„¤ì •í•˜ì—¬ ì¼ê´€ì„± ìœ ì§€)
        response = model.generate_content(
            prompt,
            generation_config={
                "response_mime_type": "application/json", 
                "temperature": 0.0
            }
        )
        
        # JSON íŒŒì‹±
        return json.loads(response.text)
        
    except Exception as e:
        print(f"âš ï¸ Error evaluating: {e}")
        # ì—ëŸ¬ ë°œìƒ ì‹œ ê¸°ë³¸ê°’ ë°˜í™˜
        return {
            "semantic_score": 0, "fluency_score": 0, 
            "cultural_score": 0, "style_score": 0, 
            "reasoning": f"Error: {str(e)}"
        }

# ==============================================================================
# 4. ë©”ì¸ ì‹¤í–‰ ë£¨í”„
# ==============================================================================
def main():
    # 1) ë°ì´í„° ë¡œë“œ
    input_file = "generation_results.json"
    print(f"ğŸ“‚ '{input_file}' ë¡œë“œ ì¤‘...")
    
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
    print(f"ğŸ“Š ì´ {len(data)}ê°œì˜ ìƒ˜í”Œì„ í‰ê°€í•©ë‹ˆë‹¤.")
    
    evaluation_results = []
    
    # 2) í‰ê°€ ìˆ˜í–‰ (tqdmìœ¼ë¡œ ì§„í–‰ë°” í‘œì‹œ)
    for entry in tqdm(data, desc="Evaluating"):
        scores = evaluate_translation(
            entry['source'], 
            entry['reference'], 
            entry['candidate']
        )
        
        # ê¸°ì¡´ ë°ì´í„°ì— ì ìˆ˜ ë³‘í•©
        evaluated_entry = {**entry, **scores}
        evaluation_results.append(evaluated_entry)
        
        # API Rate Limit ê³ ë ¤ (í•„ìš”ì‹œ ì¡°ì ˆ)
        time.sleep(1.0) 

    # 3) ê²°ê³¼ ì €ì¥ (CSV)
    output_csv = "final_evaluation_report.csv"
    df = pd.DataFrame(evaluation_results)
    df.to_csv(output_csv, index=False, encoding='utf-8-sig')
    
    print(f"\nâœ… í‰ê°€ ì™„ë£Œ! ê²°ê³¼ê°€ '{output_csv}'ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    
    # ==========================================================================
    # 5. ê²°ê³¼ ë¶„ì„ ë° ìš”ì•½ ì¶œë ¥
    # ==========================================================================
    print("\n" + "="*50)
    print("ğŸ“ˆ [ëª¨ë¸ë³„ í‰ê·  ì ìˆ˜ ë¹„êµ]")
    print("="*50)
    
    # ëª¨ë¸ë³„ ê·¸ë£¹í™” í‰ê·  ê³„ì‚°
    summary = df.groupby('model_type')[['semantic_score', 'fluency_score', 'cultural_score', 'style_score']].mean()
    print(summary)
    
    # (ì„ íƒì‚¬í•­) ì‹œê°í™”ë¥¼ ìœ„í•œ ê°„ë‹¨í•œ ë§‰ëŒ€ ê·¸ë˜í”„ ì½”ë“œ
    summary.T.plot(kind='bar', figsize=(10, 6), rot=0)
    plt.title('Performance Comparison by Metric')
    plt.ylabel('Score (0-5)')
    plt.ylim(0, 5.5)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title='Model Type')
    plt.show()

if __name__ == "__main__":
    main()
```
